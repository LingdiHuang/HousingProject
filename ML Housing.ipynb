## Checking Nulls##

```python
house = pd.read_csv('train.csv')
house.isnull().sum().sort_values(ascending=False)[:20]
```

```python
all_data_na = (house.isnull().sum() / len(house)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)[:30]
missing_data = pd.DataFrame({'Percent Missing' :all_data_na})
missing_data.head(20)
```

## Confirming Garage Variable missing ##

```python
grg = {}
garage_list = [col for col in house.columns if 'Garage' in col]

for i in garage_list:
    garage_confirm[i] = house.loc[house['GarageCars']==0][i]
#     grg = [{i:garage_confirm[i].isnull().sum() / len(garage_confirm[i]) * 100}]
    pct_missing = garage_confirm[i].isnull().sum() / len(garage_confirm[i]) * 100
    grg[i] = pct_missing

grg
```
